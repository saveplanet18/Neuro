```python
class Stack_Operation(object):

    def __init__(self,elements):
        self.elements = elements
        self. Stack = [None] * elements
        self.top = -1

    def Insertion_element(self,ele):
        if (self.top+1 == self.elements):
            print("stack is empty")
        elif (self.top == -1):
            self.top = 0
        else:
            self.top = self.top+1
        temp = self.Stack[self.top] = ele
        return temp
    def display(self):
        for i in range(len(self.Stack)):
            print(self.Stack[i])
            
```

```python
class Normal_Queue(object):

    def __init__(self,elements):
        self.elements = elements
        self.front = -1
        self.rear = -1 
        self.Queue = [None] * elements

    def Insertion_rear(self,ele):
        if (self.rear + 1) == self.elements:
            print("queue is over flow")
        elif (self.front == -1 and self.rear == -1):
            self.front = 0
            self.rear = 0
        else:
            self.rear = self.rear + 1
        temp = self.Queue[self.rear] = ele
        return temp

    def Deletion_at_front(self):
        if (self.front == -1 and self.rear == -1):
            print("Queu is empty")
        elif (self.front == self.rear):
            self.front = -1
            self.rear = -1
        else:
            temp = self.Queue[self.front]
            self.front = self.front+1
            return temp
            
    def display(self):
        if (self.front == -1 and self.rear == -1):
            print("Queue is empty")
        else:
            while (self.front != self.rear):
                print(self.Queue[self.front])
                self.front = (self.front + 1) % self.elements
            print(self.Queue[self.rear])
```
