```python
class Normal_Queue(object):

    def __init__(self,elements):
        self.elements = elements
        self.head = -1
        self.rear = -1
        self.Queue = [None] * elements
    
    def Insertion_rear(self,ele):
        if(self.rear+1) == self.elements:
            print("queue is overflow")
        elif(self.head == -1 and self.rear == -1):
            self.head = 0
            self.rear = 0
        else:
            self.rear +=1
        temp = self.Queue[self.rear] = ele
        return temp
    
    def Deletion_front(self):
        if(self.head == -1 and self.rear == -1):
            print("queue is empty")
        elif(self.head == self.rear):
            self.head = -1
            self.rear = -1
        else:
            temp = self.Queue[self.head]
            self.head = self.head+1
            return temp
    
    def showlist(self):
        if(self.head== -1 and self.rear == -1):
            print("Queue is empty")
        else:
            while(self.head != self.rear):
                print(self.Queue[self.head])
                self.head = (self.head+1) % self.elements
            print(self.Queue[self.rear])
            
```

```python
NQ = Normal_Queue(9)
NQ.Insertion_rear(108)
NQ.Insertion_rear(96)
NQ.Insertion_rear(36)
NQ.Insertion_rear(65)
NQ.Deletion_front()
NQ.Deletion_front()
NQ.showlist()
```

```python
class Dequeue(object):

    def __init__(self,elements):
        self.elements = elements
        self.head = -1
        self.rear = -1
        self.Queue = [None] * elements

    def Insertion_front(self,ele):
        if(self.head == 0 and self.rear == self.elements) or self.head == self.rear+1:
            print("Queu is overflow")
        elif(self.head == -1 and self.rear == -1):
            self.head = 0
            self.rear = 0
            temp = self.Queue[self.head] = ele
            return temp
        elif(self.head == 0):
            self.head = self.elements-1
            temp = self.Queue[self.head] = ele
            return temp
        else:
            self.head = self.head-1
            temp = self.Queue[self.head] = ele
            return temp
            
    def Insertion_rear(self,ele):
        if(self.head == 0 and self.rear == self.elements) or self.head == self.rear+1:
            print("Queue is over flow")
        elif(self.head == -1 and self.rear == -1):
            self.head = 0
            self.rear = 0
            temp = self.Queue[self.rear] = ele
            return temp
        elif(self.rear == self.elements-1):
            self.rear = 0
            temp = self.Queue[self.rear] = ele
            return temp
        else:
            self.rear = self.rear+1
            temp = self.Queue[self.rear] = ele
            return temp

    def Deletion_front(self):
        if(self.head == -1 and self.rear == -1):
            print("empty")
        elif(self.head == self.rear):
            temp = self.Queue[self.head]
            self.rear = -1
            self.head = -1
            return temp
        elif(self.head == 0):
            temp = self.Queue[self.head]
            self.head = self.head-1
            return temp
        else:
            temp = self.Queue[self.head]
            self.head = self.head+1
            return temp

    def Deletion_rear(self):
        if(self.head == -1 and self.rear == 1):
            print("Queue is empty")
        elif(self.head == self.rear):
            temp = self.Queue[self.rear]
            self.head = -1
            self.rear = -1
            return temp
        elif(self.rear == self.elements-1):
            temp = self.Queue[self.rear]
            self.rear = 0
            return temp
        else:
            temp = self.Queue[self.rear]
            self.rear = self.rear-1
            return temp

    def get_front(self):
        if(self.head == -1 and self.rear == -1):
            print("Queu is empty")
        else:
            return self.Queue[self.head]
    
    def get_rear(self):
        if(self.head == -1 and self.rear == -1):
            print("Queue is empty")
        else:
            return self.Queue[self.rear]

            
    def showlist(self):
        if(self.head == -1 and self.rear == -1):
            print("Queue is empty")
        else:
            while(self.head != self.rear):
                print(self.Queue[self.head])
                self.head = (self.head + 1) % self.elements
            print(self.Queue[self.rear])
```
