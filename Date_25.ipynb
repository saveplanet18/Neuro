```python
class Normal_Queue(object):
    def __init__(self,elements):
        self.elements = elements
        self.Queue = [None] * elements
        self.front = -1
        self.rear = -1

    def Insertion_at_rear(self,ele):
        if (self.rear + 1) == self.elements:
            print("Queue is overflow")
        elif (self.front == -1 and self.rear == -1):
            self.front = 0
            self.rear = 0
        else:
            self.rear = self.rear + 1
        temp = self.Queue[self.rear] = ele
        return temp

    def Deletion_at_front(self):
        if (self.front == -1 and self.rear == -1):
            print("Queue is empty")
        elif(self.front == self.rear):
            self.front = -1
            self.rear = -1
        else:
            temp = self.Queue[self.front]
            self.front = self.front + 1
            return temp
            

    def Traverse(self):
        if (self.rear == -1 and self.front == -1):
            print("Queue is empty")
        else:
            while (self.front != self.rear):
                print(self.Queue[self.front])
                self.front = (self.front + 1) % self.elements
            print(self.Queue[self.rear])

```

```python
class Stack(object):
    def __init__(self):
        self.Stack = []
        self.top = -1

    def Insertion_at_top(self,ele):
        if (self.top == -1):
            self.top = 0
            self.top = self.top + 1
            temp = self.Stack.append(ele)
            return temp
        else:
            print("emtpy")

    def Deletion_at_top(self):
        temp = self.Stack.pop(1)
        return temp
    
    
    def Traverse(self):
        temp = self.Stack
        for i in range(len(temp)):
            print(temp[i])
```
